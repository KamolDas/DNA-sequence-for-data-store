```python
import matplotlib.pyplot as plt

def plot_string_indices(input_string, slice_start=None, slice_end=None):
    # Slice the string if start and end are provided
    if slice_start is not None and slice_end is not None:  # slice index given so its will work
        input_string = input_string[slice_start:slice_end]

    # Define a mapping for the input string characters to indices
    char_to_index = {char: idx for idx, char in enumerate(sorted(set(input_string)))}

    # Convert the input string to indices
    indices = [char_to_index[char] for char in input_string]

    # Plotting
    plt.figure(figsize=(15, 5))
    plt.plot(range(len(input_string)), indices, 'o-', label='Character Indices')
    plt.xticks(range(len(input_string)), list(input_string))
    plt.xlabel('Character of str')
    plt.ylabel('Index of str')
    plt.title('  Plot of Characters and Their Indices')
    plt.suptitle('Ploting of Characters and Their index value(indices)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
input_string = "ABCDDAj gsdsjdf fcCBD ACBDjdfhCABCC"   # gap index as '0'
plot_string_indices(input_string)

# Example with slicing
plot_string_indices(input_string, slice_start=6, slice_end=26)
```
